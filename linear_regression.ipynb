```python
def gradient_descent(m_now, b_now, X, Y, lr):
    m_gradient = 0
    b_gradient = 0

    n = len(X)

    for i in range(n):
        x = X[i]
        y = Y[i]

        m_gradient += -(2 / n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2 / n) * (y - (m_now * x + b_now))

    m = m_now - m_gradient * lr
    b = b_now - b_gradient * lr

    return m, b
```
